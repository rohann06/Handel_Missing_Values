```python
import pandas as pd

data = pd.read_csv("Houseprice.csv", usecols=['BsmtQual', 'FireplaceQu', 'GarageType', 'SalePrice'])
data.head()
```

```python
data.isnull().sum()
```

```python
import numpy as np

data['BsmtQual_var'] = np.where(data['BsmtQual'].isnull(),1,0)
```

```python
data.head()
```

```python
m_f_r = data['BsmtQual'].value_counts().index[0]
```

```python
data['BsmtQual'].fillna(m_f_r,inplace=True)
data.head()
```
