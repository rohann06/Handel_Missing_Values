# Frequent categories imputation

    -> Replace NAN With Most  Frequent Values
### CONS
    -> Can't Use On Big Amount Of NAN values

```python
import pandas as pd

data = pd.read_csv("Houseprice.csv", usecols=['BsmtQual', 'FireplaceQu', 'GarageType', 'SalePrice'])
data.head()
```

```python
data.isnull().sum()
```

```python
data.isnull().mean().sort_values(ascending=True)
```

### Compute Veryfrequent in every feture

```python
data['BsmtQual'].value_counts().plot.bar()
```

```python
data['GarageType'].value_counts().plot.bar()
```

```python
data['FireplaceQu'].value_counts().plot.bar()
```

```python
### Replacing The NAN Values

def fill_na(data,variable):
    m_f_r = data[variable].value_counts().index[0]
    data[variable].fillna(m_f_r,inplace = True)
```

```python
for values in ['BsmtQual','GarageType', 'FireplaceQu']:
    fill_na(data,values)
```
